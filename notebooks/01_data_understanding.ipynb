```python
from pathlib import Path
import sys

PROJECT_ROOT = Path().resolve().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import pandas as pd
from retailpulse.utils import explore_dataframe
```

```python
orders = pd.read_csv("../data/raw/olist_orders_dataset.csv")
customers = pd.read_csv("../data/raw/olist_customers_dataset.csv")
groups = pd.read_csv("../data/raw/olist_geolocation_dataset.csv")
items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
payments = pd.read_csv("../data/raw/olist_order_payments_dataset.csv")
reviews = pd.read_csv("../data/raw/olist_order_reviews_dataset.csv")
products = pd.read_csv("../data/raw/olist_products_dataset.csv")
sellers = pd.read_csv("../data/raw/olist_sellers_dataset.csv")
categories = pd.read_csv("../data/raw/product_category_name_translation.csv")
```

```python
explore_dataframe(orders, "Orders Dataset")
```

### Initial Observations

- The dataset contains **99,441 orders** and **8 columns**.
- The identifier columns (`order_id` and `customer_id`) contain no missing values.
- Three timestamp columns contain missing values:
  - `order_approved_at` (160)
  - `order_delivered_carrier_date` (1,783)
  - `order_delivered_customer_date` (2,965)
- All timestamp columns are currently stored as `object` and should be converted to `datetime`.
- Further investigation is required to determine whether the missing timestamp values are expected (e.g., canceled orders) or represent data quality issues.

```python
explore_dataframe(customers, "Customers Dataset")
```

### Initial Observations

- The dataset contains **99,441** unique customer records.
- No missing values or duplicate rows were found.
- Customer location information is available through city, state, and ZIP code.
- This dataset will be used to analyze customer distribution across different regions.

```python
explore_dataframe(groups, "Geolocation Dataset")
```

### Initial Observations

- The dataset contains more than **1 million** location records.
- No missing values were found.
- A large number of duplicate rows (**261,831**) were detected.
- Duplicate records should be investigated before deciding whether they should be removed.
- This dataset can be used for geographical analysis and customer location visualization.

```python
explore_dataframe(items, "Order Items Dataset")
```

### Initial Observations

- The dataset contains **112,650** order item records.
- No missing values or duplicate rows were found.
- Each record represents a product included in an order.
- Product price and shipping cost are available for profitability and sales analysis.

```python
explore_dataframe(payments, "Order Payments Dataset")
```

### Initial Observations

- The dataset contains **103,886** payment transactions.
- No missing values or duplicate rows were found.
- Multiple payment methods and installment information are available.
- This dataset can be used to analyze customer payment preferences.

```python
explore_dataframe(reviews, "Order Reviews Dataset")
```

### Initial Observations

- The dataset contains **99,224** customer reviews.
- Missing values are mainly found in review titles and review comments.
- Missing review text is expected because customers may submit only a rating without leaving a written comment.
- Review scores will be useful for measuring customer satisfaction.

```python
explore_dataframe(products, "Products Dataset")
```

### Initial Observations

- The dataset contains **32,951** products.
- Missing values exist in product category and product attribute columns.
- Product dimensions and weight contain a few missing values.
- Product information will support category-level sales analysis.

```python
explore_dataframe(sellers, "Sellers Dataset")
```

### Initial Observations

- The dataset contains **3,095** sellers.
- No missing values or duplicate rows were found.
- Seller location information is available through city, state, and ZIP code.
- This dataset will be useful for analyzing seller distribution and performance.

```python
explore_dataframe(categories , "Product Categories Dataset")
```

### Initial Observations

- The dataset contains **71** product category mappings.
- No missing values or duplicate records were found.
- It provides a translation from Portuguese product category names to English.
- This dataset will be merged with the Products dataset to improve category readability in the analysis and dashboard.
