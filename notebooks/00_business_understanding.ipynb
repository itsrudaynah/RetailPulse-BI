# Business Understanding

## Project Overview

RetailPulse BI is a Business Intelligence project that analyzes the Brazilian E-Commerce Public Dataset (Olist). The objective is to understand customer purchasing behavior, product performance, payment patterns, and delivery efficiency in order to generate actionable business insights.

## Business Objectives

The main objectives of this project are:

- Understand customer purchasing behavior.
- Identify the best-selling products and categories.
- Analyze delivery performance.
- Evaluate customer satisfaction.
- Explore payment behavior.
- Build an interactive Business Intelligence dashboard.

## Business Questions

1. What are the peak purchasing hours during the day?

2. Which days of the week have the highest number of orders?

3. Which months generate the highest sales?

4. Which product categories generate the highest revenue?

5. Which payment methods are most commonly used?

6. How long does order approval usually take?

7. How long does shipping take?

8. Are orders delivered before or after the estimated delivery date?

9. Which states and cities generate the highest sales?

10. Does shipping cost affect customer review scores?

11. Which sellers generate the highest revenue?

12. Which products receive the highest customer ratings?
